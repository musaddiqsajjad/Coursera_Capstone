```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import os
import time
import json
```

Reading GeoJSON files

```python
with open(r"json\regions.json", 'r', encoding='utf8', errors='ignore') as file:
    regions = json.load(file)

with open(r"json\cities.json", 'r', encoding='utf8', errors='ignore') as file:
    cities = json.load(file)
    
with open(r"json\districts.json", 'r', encoding='utf8', errors='ignore') as file:
    districts = json.load(file)
```

Getting city_id of Al Khobar and Dammam

```python
for city in cities:
    if "Khobar" in city["name_en"]:
        print(str(city["city_id"]) + " " + city["name_en"])
    if "Dammam" in city["name_en"]:
        print(str(city["city_id"]) + " " + city["name_en"])
```

Getting distrcits in Al Khobar and Dammam

```python
khobar_districts = []
dammam_districts = []

for district in districts:
    if district["city_id"] == 31:
        khobar_districts.append(district)
        
for district in districts:
    if district["city_id"] == 13:
        dammam_districts.append(district)
```

```python
khobar = pd.DataFrame(khobar_districts)
```

```python
geojson = {
    "type": "FeatureCollection",
    "features": [{
        "type": "Feature",
        "geometry": {
            "type": "Polygon",
            "coordinates": khobar_districts[1]["boundaries"]},
        "properties": {
            "district_id": khobar_districts[1]["district_id"],
            "city_id": khobar_districts[1]["city_id"],
            "region_id": khobar_districts[1]["region_id"],
            "name_ar": khobar_districts[1]["name_ar"],
            "name_en": khobar_districts[1]["name_en"]}
    }
    ]
}

geojson = json.dumps(geojson)
geojson
```

```python
khobar_data = khobar[["name_en", "region_id", "district_id"]]
khobar_data
```

```python
# create a plain map
map = folium.Map(location=[26.2172,50.1971], zoom_start=15)

folium.GeoJson(geojson).add_to(map)

# display map
map
```
